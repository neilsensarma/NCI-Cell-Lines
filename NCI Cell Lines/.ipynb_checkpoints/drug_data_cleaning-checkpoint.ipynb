```python
df = pd.read_excel('C:\\GaTech\\Ph. D\\Research\\Datasets\\New\\DTP_NCI60_RAW\\output\\DTP_NCI60_RAW.xlsx')
df = df.iloc[8:,]
df.reset_index(inplace=True, drop=True)
```

```python
df.head(2)
```

```python
df.columns = df.iloc[0]
```

```python
df.head(5)
```

```python
df = df.iloc[1:]
drug_names = pd.Series(df['Drug name'])
df.drop(columns = ['NSC #', 'FDA status','Experiment name','Mechanism of action', 'PubChem SID', 'Total probes', 'Total after quality control', 'Failure reason'], inplace=True)
cols = list(df.columns)[2:]
df = df[cols].astype('float', errors='ignore')
df.replace(to_replace = 'na', value='0.0', inplace=True)
df['Drugs'] = drug_names
```

```python
drug_names
```

```python
df
```

```python
df.reset_index(drop = True, inplace = True)
```

```python
df.head()
```

```python
trial = df.astype('float', errors = 'ignore')
```

```python
trial.head()
```

```python
trial2 = trial.groupby('Drugs', sort=False).mean().reset_index()
```

```python
trial2.head()
```

```python
'Gemcitabine' in list(trial2.Drugs)
```

```python
working_data = trial2[trial2['Drugs'] == 'Gefitinib']
working_data.head()
```

```python
working_data.reset_index(inplace=True, drop=True)
working_data.head()
```

```python
wd = working_data.T
```

```python
wd.index.names = ['hello']
wd
```

```python
wd.reset_index(inplace=True)
wd.head()
```

```python
wd.columns = wd.iloc[0]
wd.rename(columns = {'Drugs': 'Cell names'})
wd = wd.iloc[1:, :]
wd.head(2)
```

```python
wd.rename(columns = {'Drugs': 'Cell names'}, inplace=True)
wd.head(2)
```

```python
wd.reset_index(drop=True, inplace=True)
```

```python
wd.head(5)
```

```python
names = []
cells = list(wd['Cell names'])
for i in range(len(cells)):
    names.append([''.join(a.split('-')) for a in [x for x in cells[i].split(':')]][1])
```

```python
names
```

```python
cell_names = [''.join(a.split(' ')) for a in names]
cell_names
```

```python
wd['cells'] = cell_names
wd.head(2)
```

```python
wd.to_csv('Gefitinib_NCI.csv', index=False)
```

```python
metabolomic.rename(columns = {'names': 'cells'}, inplace=True)
wd2 = metabolomic.merge(wd, on='cells')
wd2.shape
```
