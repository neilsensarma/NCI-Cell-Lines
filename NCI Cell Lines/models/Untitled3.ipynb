```python
import numpy as np
import pandas as pd
import umap
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:\\Users\\nilav\\NCI Cell Lines\\Gemcitabine_metabolic_NCI.csv')
data.head(2)
```

```python
drug = data.pop('Gemcitabine')
drug.head()
```

```python
X = data.select_dtypes('float64')
X.head(2)
```

```python
X.fillna(0.0, inplace=True)
```

```python
labels = []
drug_mean = drug.mean()
for i in drug:
    if i >= drug_mean:
        labels.append('N')
    else:
        labels.append('R')
y = pd.Series(labels)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_transformed = pd.DataFrame(pca.fit_transform(X))
```

```python
X_transformed
```

```python
fig, ax = plt.subplots(figsize = (5,5))
plt.scatter(x = X_transformed.iloc[:, :1], y = X_transformed.iloc[:, 1:2], c=y, cmap='Spectral', alpha = 1.0)
ax.set_title('Comparison of metabolite profiles for NCI cell lines in response to Gemcitabine')
ax.set_xlabel('Component 0')
ax.set_ylabel('Component 1')
ax.legend(y)
plt.show()
```
